```python
import torch
from ogb.nodeproppred import PygNodePropPredDataset
```

```python
# Load the OGBN-Arxiv dataset
dataset = PygNodePropPredDataset(name='ogbn-arxiv')

# Access the data
data = dataset[0]

# Print dataset information
print(f'Dataset: {dataset.name}')
print(f'Number of nodes: {data.num_nodes}')
print(f'Number of edges: {data.num_edges}')
print(f'Node feature matrix shape: {data.x.shape}')
print(f'Node labels shape: {data.y.shape}')
```

```python
# Print a few details about the dataset
print(f'Node features (first 5): {data.x[:5]}')
print(f'Node labels (first 5): {data.y[:5]}')
print(f'Edge index (first 5 edges): {data.edge_index[:, :5]}')
```

```python
split_idx = dataset.get_idx_split()
train_idx = split_idx['train']
val_idx = split_idx['valid']
test_idx = split_idx['test']

print(f'Train indices (first 5): {train_idx[:5]}')
print(f'Validation indices (first 5): {val_idx[:5]}')
print(f'Test indices (first 5): {test_idx[:5]}')
```
